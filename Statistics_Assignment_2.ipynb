#Q1. 1. What is hypothesis testing in statistics?
  - A systematic method for making inferences about a population based on sample data by assessing whether observed effects could reasonably occur under a specific assumption.
#Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H₀) states no effect or difference, while the alternative hypothesis (H₁) proposes the effect or difference you're testing for.
#Q3. What is the significance level in hypothesis testing, and why is it important?
  - The significance level (α) is the threshold probability (commonly 0.05) for rejecting H₀, balancing the risk of a false positive.
#Q4. What does a P-value represent in hypothesis testing?
  - The P-value is the probability of observing data as extreme or more extreme than what was observed, assuming H₀ is true.
#Q5. How do you interpret the P-value in hypothesis testing?
  - A P-value less than α leads to rejecting H₀; a P-value greater than α means you fail to reject H₀.
#Q6. What are Type 1 and Type 2 errors in hypothesis testing?
  - A Type 1 error is falsely rejecting a true H₀ (false positive), and a Type 2 error is failing to reject a false H₀ (false negative).
#Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - A one-tailed test checks for an effect in one direction, while a two-tailed test checks for an effect in either direction.
#Q8. What is the Z-test, and when is it used in hypothesis testing?
  - A Z-test assesses whether a sample mean significantly differs from a known population mean when the population standard deviation is known and samples are large.
#Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - The Z-score is (sample mean - population mean) divided by (population standard deviation /√n), representing how many standard errors the sample mean is from the population mean.
#Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The T-distribution is a symmetric, bell-shaped distribution with heavier tails, used when the population standard deviation is unknown and the sample size is small.
#Q11. What is the difference between a Z-test and a T-test?
  - A Z-test uses known population standard deviation, whereas a T-test uses an estimated standard deviation from the sample.
#Q12. What is the T-test, and how is it used in hypothesis testing?
  - A T-test evaluates whether the means of one or two groups significantly differ when population variance is unknown.
#Q13. What is the relationship between Z-test and T-test in hypothesis testing?
  - The T-test converges to the Z-test as sample size increases and the sample standard deviation approaches the population standard deviation.
#Q14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval provides a range of plausible values for a population parameter with a specified confidence level (e.g., 95%).
#Q15. What is the margin of error, and how does it affect the confidence interval?
  - The margin of error is half the width of the confidence interval, reflecting the maximum expected difference between the sample estimate and the true population parameter.
#Q16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes' Theorem updates the probability of a hypothesis based on new evidence, combining prior beliefs with data likelihood.
#Q17. What is the Chi-square distribution, and when is it used?
  - The Chi-square distribution is a family of right-skewed distributions used for inference about variances and categorical data.
#Q18. What is the Chi-square goodness of fit test, and how is it applied?
  - The Chi-square goodness of fit test assesses whether observed categorical frequencies significantly deviate from expected frequencies under a specified distribution.
#Q19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is a right-skewed distribution used to compare variances or test multiple means (e.g., ANOVA).
#Q20. What is an ANOVA test, and what are its assumptions?
  - ANOVA is a method to compare three or more group means assuming independent samples, normality within groups, and equal variances.
#Q21. What are the different types of ANOVA tests?
  - The main types include one-way ANOVA (one factor), two-way ANOVA (two factors), and repeated measures ANOVA.
#Q22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test evaluates whether variances or multiple means are equal by comparing two variances through an F-statistic.

```python
#Q12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal():
    x = np.linspace(-4, 4, 400)
    y = norm.pdf(x, 0, 1)
    plt.plot(x, y, label="N(0,1)")
    plt.fill_between(x, y, alpha=0.2)
    plt.title("Standard Normal PDF")
    plt.xlabel("x"); plt.ylabel("Density")
    plt.legend()
    plt.show()

plot_standard_normal()
```

```python
#Q14. Write a Python program to calculate the 2-score for a given data point and compare it to a standard normal distribution.
import numpy as np
from scipy.stats import norm

def z_score(x, data):
    mu, sigma = np.mean(data), np.std(data, ddof=1)
    return (x - mu) / sigma

data = np.random.normal(100, 15, size=100)
z = z_score(120, data)
print("Z-score:", z, "=> P(X>Z)=", 1 - norm.cdf(z))
```

```python
#Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats.weightstats import ztest

data = np.random.normal(50, 5, size=100)
stat, p = ztest(data, value=52)
alpha = 0.05
crit = norm.ppf(1 - alpha/2)

x = np.linspace(stat-4, stat+4, 200)
plt.plot(x, norm.pdf(x))
plt.fill_between(x, norm.pdf(x), where=(x>crit)|(x<-crit), color='red', alpha=0.3)
plt.axvline(stat, color='black', label=f"Z={stat:.2f}")
plt.legend(); plt.show()
print("P‑value:", p)
```

```python
#Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu0, mu1, sigma, n, alpha = 0, 1, 1, 30, 0.05
se, crit = sigma/np.sqrt(n), norm.ppf(1-alpha)
β = norm.cdf(crit, loc=mu1/se)  # Type II error
plt.plot(x := np.linspace(-4, 6, 300), norm.pdf(x, mu0/se), label="H0")
plt.plot(x, norm.pdf(x, mu1/se), label="H1")
plt.axvline(crit, color='k')
plt.fill_between(x, norm.pdf(x, mu0/se), where=(x>crit), color='red', alpha=0.3, label="α")
plt.fill_between(x, norm.pdf(x, mu1/se), where=(x<crit), color='blue', alpha=0.3, label="β")
plt.legend(); plt.show()
print("Type I (α):", alpha, "Type II (β):", β)
```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import norm

def ci_mean(data, confidence=0.95):
    mu, se = np.mean(data), np.std(data, ddof=1)/np.sqrt(len(data))
    h = se * norm.ppf((1+confidence)/2)
    return mu - h, mu + h

data = np.random.normal(50, 8, size=40)
print("95% CI:", ci_mean(data))
```

```python
#Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import norm

def ci_mean(data, confidence=0.95):
    mu, se = np.mean(data), np.std(data, ddof=1)/np.sqrt(len(data))
    h = se * norm.ppf((1+confidence)/2)
    return mu - h, mu + h

data = np.random.normal(50, 8, size=40)
print("95% CI:", ci_mean(data))
```

```python
#Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np
from scipy.stats import chi2_contingency

observed = np.array([
    [30, 10],
    [20, 40]
])
print("Observed frequencies:\n", observed)

chi2, p_value, dof, expected = chi2_contingency(observed)

print(f"\nChi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p_value:.4f}")
print(f"Expected frequencies under H₀:\n{expected}")
```
